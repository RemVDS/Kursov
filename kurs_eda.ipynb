```python
columns_with_zeros = df.columns[(df == 0).all(axis=0)] # в ходе исследования заметили, что есть признаки равные только 0

# Показать результат
display(columns_with_zeros)
```

```python
columns_drop = ['NumRadicalElectrons', 'SMR_VSA8', 'SlogP_VSA9', 'fr_N_O', 'fr_SH',
       'fr_azide', 'fr_barbitur', 'fr_benzodiazepine', 'fr_diazo',
       'fr_dihydropyridine', 'fr_isocyan', 'fr_isothiocyan', 'fr_lactam',
       'fr_nitroso', 'fr_phos_acid', 'fr_phos_ester', 'fr_prisulfonamd',
       'fr_thiocyan']
```

```python
df = df.drop(columns_drop, axis=1) # удалим признаки, равные 0
```

```python
columns_with_zeros = df.columns[(df == 0).all(axis=0)] # проверим результат удаления

# Показать результат
print(columns_with_zeros)
```

```python
constant_cols = [col for col in df.columns if df[col].nunique() == 1]  # Найдём и удаляем признаки, не содержащие разнообразия (константные столбцы)
df.drop(columns=constant_cols, inplace=True)
```

```python
# Основные статистики для ключевых переменных
key_columns = ["IC50, mM", "CC50, mM", "SI"]
display(df[key_columns].describe())
```

```python
# Гистограммы для IC50, CC50, SI
plt.figure(figsize=(15, 5))
for i, col in enumerate(key_columns, 1):
    plt.subplot(1, 3, i)
    sns.histplot(df[col], kde=True)
    plt.title(f"Распределение {col}")
plt.tight_layout()
plt.show()
```

```python
# Матрица корреляций для ключевых переменных
corr_matrix = df.corr(numeric_only=True)
plt.figure(figsize=(12, 12))
sns.heatmap(corr_matrix[key_columns].T, annot=True, cmap="coolwarm")
plt.title("Корреляция между IC50, CC50, SI")
plt.show()
```

```python
sns.scatterplot(data=df, x="IC50, mM", y="CC50, mM", hue="SI")
plt.title("IC50 vs CC50 (цвет — SI)")
plt.show()
```

```python
plt.figure(figsize=(10, 5)) # проверка на выбросы
sns.boxplot(data=df[key_columns])
plt.title("Boxplot для IC50, CC50, SI")
plt.xticks(rotation=45)
plt.show()
```

```python
# Визуализация с применением логарифмирования для лучшего отображения данных
target_cols = ["IC50, mM", "CC50, mM", "SI"]
log_df = df[target_cols].apply(lambda x: np.log10(x + 1e-6))
sns.pairplot(log_df)
plt.show()

# Корреляционная матрица
corr = df[target_cols].corr(method='spearman')
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Корреляция между IC50, CC50 и SI")
plt.show()
```

**Выводы по EDA**
   - Выбросы, которые мы наблюдаем не известной мне природы, так как работаем с научным датасетом удалять их не будем, так как в полне им может  быть объяснение химической природы. 
   - Признаки **IC50 и CC50** имеют умеренную положительную корреляцию (0.62).  
   - Признак **SI** (CC50/IC50) отрицательнон коррелирует с IC50 (–0.64) и положительно — с CC50 (0.09).  
   - Ни один признак не имеет сильной  корреляции с целевыми переменными, что подсказывает в дальнейшем обратиться к сложным моделям.  
   - Логарифмическое отображение подчеркивает кластеры “высокоактивных” (низкий IC50) и “низкотоксичных” (высокий CC50) соединений.
